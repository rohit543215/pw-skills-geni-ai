## 1. What is hypothesis testing in statistics?

### hypothesis testing in statistics is a method used to make decisions or inferences about a population based on a sample data. it helps you determine whether there is enough evidence to support a specific claim or belief about a population parameter. 

## 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

### üìå Null Hypothesis (H‚ÇÄ):
- The null hypothesis is a statement that there is no effect, no difference,     or no relationship in the population.
- It is the default assumption we test against.
- We assume the null hypothesis is true unless we have strong evidence to      reject it.
- üî∏Example:
- H‚ÇÄ: the average height of students= 170cm (Œº = 170)
- H‚ÇÄ: A new drug has no effect on blood pressure.

### üìå Alternative Hypothesis (H‚ÇÅ or Ha):
- The alternative hypothesis is what you want to prove or support.
- it suggests that there is an effect, a difference, or a relationship.
- if the evidence is dtrong enough to reject H‚ÇÄ, we accept H‚ÇÅ.
- üî∏ Example:
- H‚ÇÅ: The average height of students ‚â† 170 cm (Œº ‚â† 170)
- H‚ÇÅ: The new drug does affect blood pressure.

## 3.What is the significance level in hypothesis testing, and why is it important?

### üìå Definition:
The significance level (denoted by Œ±, alpha) is the threshold for deciding whether a result is statistically significant.
It represents the maximum probability of making a Type I error ‚Äî that is, rejecting the null hypothesis (H‚ÇÄ) when it is actually true.

### üß† Why Is It Important?
üß† Why Is It Important
- 1.Sets the standard for decision-making:
   - It helps you decide whether your test result is due to random chance or   real effect.
  -You compare the p-value to Œ±:
    - If p-value < Œ± ‚Üí reject H‚ÇÄ (significant result).
    - If p-value ‚â• Œ± ‚Üí fail to reject H‚ÇÄ (not significant).
- 2.Controls risk:
  - By choosing Œ±, you control how willing you are to make a false positive   (Type I error).
  - For example, with Œ± = 0.05, you're accepting a 5% chance of wrongly  rejecting a true null hypothesis.
- 3.Balancing errors:
  - Lower Œ± reduces the chance of a Type I error but increases the chance of a Type II error (failing to detect a real effect).



## 4.What does a p-value represent in hypothesis testing.?

### üìå Definition:
The p-value is the probability of observing the sample results (or something more extreme) assuming the null hypothesis (H‚ÇÄ) is true.

### üí° In Simple Words:
- It answers the question:
  "If there really is no effect (H‚ÇÄ is true), how likely is it that we would see these results?"
- A small p-value means that the observed data is unlikely under H‚ÇÄ, so we have evidence against H‚ÇÄ.

## 5. How do you interpret the p-value in hypothesis testing?

### üéØ In Simple Terms:
- Small p-value (e.g., 0.01):
  There‚Äôs only a 1% chance that the observed results are due to random chance if H‚ÇÄ is true ‚Üí strong reason to reject H‚ÇÄ.
- Large p-value (e.g., 0.6):
 There's a 60% chance the results are just due to randomness ‚Üí not enough evidence to reject H‚ÇÄ.

## 6.What are type-1 and type-2 errors in hypothesis testing? 

### üìå 1. Type I Error (False Positive):
     - Definition: Rejecting the null hypothesis (H‚ÇÄ) when it is actually   true.
     - In Simple Terms: You think there is an effect, but there isn‚Äôt.

### üìå 2. Type II Error (False Negative):
- Definition: Failing to reject the null hypothesis (H‚ÇÄ) when the alternative hypothesis (H‚ÇÅ) is actually true.
- In Simple Terms: You think there‚Äôs no effect, but there actually is.

## 7.What is the difference between a one-tailed and two-tailed test in hypothesis testing?

### üìå 1. One-Tailed Test:
- Used when you are testing for an effect in one specific direction.
- You are only interested in either increase or decrease, not both.

üî∏ Example:
- You want to test if a new teaching method improves test scores (not just changes them).

- H‚ÇÄ: Œº ‚â§ 70
- H‚ÇÅ: Œº > 70 (one-tailed to the right)


### üìå 2. Two-Tailed Test:
- Used when you are testing for an effect in either direction.
- You care about any difference, not whether it's specifically higher or lower.

üî∏ Example:
You want to test if a machine's output differs from the standard 100g (could be more or less).
- H‚ÇÄ: Œº = 100
- H‚ÇÅ: Œº ‚â† 100

## 9.How do you calculate the Z-score, and what does it represent in hypothesis testing? 

### üìå What is a Z-score?
A Z-score tells you how many standard deviations a data point (or sample mean) is from the population mean.

In hypothesis testing, it helps you decide whether to reject the null hypothesis (H‚ÇÄ) by comparing the Z-score to critical values from the standard normal distribution.

üìå What It Represents:
- A Z-score of 0 ‚Üí Sample mean = population mean (no difference)
- A positive Z-score ‚Üí Sample mean is above the population mean
- A negative Z-score ‚Üí Sample mean is below the population mean
- The larger the absolute value of Z, the more extreme the result

The T-distribution (Student's t-distribution) is a probability distribution similar to the normal distribution but with heavier (thicker) tails.

These heavier tails account for the extra uncertainty that comes when:
- Sample size is small
- Population standard deviation (œÉ) is unknown

## 1. performing a Z-test to compare a sample mean with a known population mean:

```python
from scipy.stats import norm
import numpy as np

# Sample data
data = [104, 102, 98, 101, 99, 105, 100, 103, 97, 106]
sample_mean = np.mean(data)
population_mean = 100
population_std = 5
n = len(data)

# Z-test
z = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p = 2 * (1 - norm.cdf(abs(z)))

print("Z-score:", z)
print("P-value:", p)
```

## 4.Perform a two-tailed Z-test using python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample data
data = [104, 102, 98, 101, 99, 105, 100, 103, 97, 106]
mean = np.mean(data)
n = len(data)
pop_mean = 100
pop_std = 5

# Z-test
z = (mean - pop_mean) / (pop_std / np.sqrt(n))
p = 2 * (1 - norm.cdf(abs(z)))
print("Z-score:", z)
print("P-value:", p)

# Plot
x = np.linspace(-4, 4, 300)
y = norm.pdf(x)
z_crit = norm.ppf(0.975)

plt.plot(x, y)
plt.axvline(z, color='red', label=f'z = {z:.2f}')
plt.axvline(-z_crit, color='gray', linestyle='--')
plt.axvline(z_crit, color='gray', linestyle='--')
plt.fill_between(x, y, where=(x <= -z_crit) | (x >= z_crit), color='red', alpha=0.2)
plt.title("Two-tailed Z-test")
plt.legend()
plt.show()
```

## 5.Create a python function that calculates and visualizes type-1 and type-2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2(mu0=100, mu1=105, sigma=10, alpha=0.05, n=30):
    se = sigma / np.sqrt(n)
    z_crit = norm.ppf(1 - alpha/2)

    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 500)
    y0 = norm.pdf(x, mu0, se)  # Null distribution
    y1 = norm.pdf(x, mu1, se)  # Alternative distribution

    plt.plot(x, y0, label="H‚ÇÄ distribution", color='blue')
    plt.plot(x, y1, label="H‚ÇÅ distribution", color='green')

    # Critical values
    cv_left = mu0 - z_crit * se
    cv_right = mu0 + z_crit * se

    # Shade Type-I error (Œ±)
    plt.fill_between(x, y0, where=(x < cv_left) | (x > cv_right), color='red', alpha=0.3, label='Type-I Error (Œ±)')

    # Shade Type-II error (Œ≤)
    plt.fill_between(x, y1, where=(x > cv_left) & (x < cv_right), color='orange', alpha=0.3, label='Type-II Error (Œ≤)')

    plt.legend()
    plt.title("Type-I and Type-II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Example use
plot_type1_type2()
```

## 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance. give codes for this one by one simple and small.

```python
import numpy as np
from scipy.stats import t

def ci_mean(data, conf=0.95):
    data = np.asarray(data)
    n, mean, s = len(data), data.mean(), data.std(ddof=1)
    se = s / np.sqrt(n)
    h  = t.ppf((1 + conf) / 2, n-1) * se
    return mean - h, mean + h

# Example
sample = [5.1, 4.9, 5.0, 5.2, 4.8]
print("95% CI:", ci_mean(sample))
```

## 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
# P(Disease|Positive) = P(Positive|Disease)*P(Disease) / P(Positive)

p_disease = 0.01
p_pos_given_disease = 0.99
p_pos_given_no_disease = 0.05
p_no_disease = 1 - p_disease

# Total probability of positive
p_positive = p_pos_given_disease * p_disease + p_pos_given_no_disease * p_no_disease

# Bayes' rule
posterior = (p_pos_given_disease * p_disease) / p_positive
print("P(Disease | Positive Test):", round(posterior, 3))
```

## 14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.


```python
from scipy.stats import chisquare

# Observed and expected frequencies
observed = [18, 22, 20]
expected = [20, 20, 20]  # uniform distribution

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Display results
print("Chi-square Statistic:", round(chi2_stat, 2))
print("P-value:", round(p_value, 4))

# Interpretation
if p_value < 0.05:
    print("‚Üí Reject H‚ÇÄ (Observed distribution differs significantly)")
else:
    print("‚Üí Fail to reject H‚ÇÄ (No significant difference)")
```

## 16. Implement an F-test using Python to compare the variances of two random samples.


```python
import numpy as np
from scipy.stats import f

# Two samples
sample1 = [10, 12, 9, 11, 13]
sample2 = [7, 8, 6, 9, 7]

# Calculate variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# F-statistic
f_stat = var1 / var2
dof1 = len(sample1) - 1
dof2 = len(sample2) - 1

# One-tailed p-value
p_value = 1 - f.cdf(f_stat, dof1, dof2)

print("F-statistic:", round(f_stat, 3))
print("P-value:", round(p_value, 4))

# Interpretation
if p_value < 0.05:
    print("‚Üí Reject H‚ÇÄ (Variances are significantly different)")
else:
    print("‚Üí Fail to reject H‚ÇÄ (No significant difference in variances)")
```

## 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups):
    # 1. Normality check (Shapiro-Wilk test)
    for i, group in enumerate(groups, start=1):
        stat, p = shapiro(group)
        print(f"Group {i} Normality p-value: {p:.4f}")
        if p > 0.05:
            print(" ‚Üí Data looks normal.")
        else:
            print(" ‚Üí Data may not be normal.")

    # 2. Equal variances (Levene‚Äôs test)
    stat, p = levene(*groups)
    print(f"\nLevene‚Äôs Test (Equal Variance) p-value: {p:.4f}")
    if p > 0.05:
        print(" ‚Üí Variances are likely equal.")
    else:
        print(" ‚Üí Variances may not be equal.")
    
    # 3. Independence - Visual check using line plot (assumes ordered data)
    for i, group in enumerate(groups, start=1):
        plt.plot(group, label=f'Group {i}')
    plt.title("Line Plot to Visually Check Independence")
    plt.xlabel("Index")
    plt.ylabel("Values")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
g1 = [23, 25, 27, 22, 24]
g2 = [31, 33, 30, 29, 32]
g3 = [18, 20, 19, 21, 17]

check_anova_assumptions(g1, g2, g3)
```

## 20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.


```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Data: Scores by 'Diet' and 'Exercise'
data = {
    'Score': [80, 82, 84, 78, 76, 75, 90, 88, 85, 70, 72, 73],
    'Diet': ['High', 'High', 'High', 'Low', 'Low', 'Low', 'High', 'High', 'High', 'Low', 'Low', 'Low'],
    'Exercise': ['Yes', 'Yes', 'No', 'Yes', 'Yes', 'No', 'No', 'No', 'Yes', 'No', 'Yes', 'Yes']
}

df = pd.DataFrame(data)

# Two-way ANOVA
model = ols('Score ~ C(Diet) + C(Exercise) + C(Diet):C(Exercise)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("\nTwo-Way ANOVA Results:")
print(anova_table)

# Visualization: Interaction plot
sns.pointplot(x='Diet', y='Score', hue='Exercise', data=df, markers=["o", "s"], linestyles=["-", "--"])
plt.title("Interaction Between Diet and Exercise")
plt.grid(True)
plt.show()
```

## 24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.


```python
import numpy as np
from scipy.stats import chi2

# Sample data
data = [10, 12, 9, 11, 13, 10, 11, 12]
n = len(data)
sample_var = np.var(data, ddof=1)

# Hypothesized population variance (e.g., H0: œÉ¬≤ = 4)
sigma_squared_0 = 4

# Chi-square statistic
chi2_stat = (n - 1) * sample_var / sigma_squared_0

# Degrees of freedom
df = n - 1

# p-value (two-tailed)
p_value = 2 * min(chi2.cdf(chi2_stat, df), 1 - chi2.cdf(chi2_stat, df))

# Output
print(f"Sample Variance = {round(sample_var, 2)}")
print(f"Chi-square Stat = {round(chi2_stat, 2)}")
print(f"p-value = {round(p_value, 4)}")

# Interpretation
if p_value < 0.05:
    print("‚Üí Reject H‚ÇÄ: Population variance is significantly different from", sigma_squared_0)
else:
    print("‚Üí Fail to reject H‚ÇÄ: No evidence that variance differs from", sigma_squared_0)
```

## 25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.


```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Example:
# Group A: 40 successes out of 100
# Group B: 30 successes out of 90
successes = np.array([40, 30])
samples = np.array([100, 90])

# Perform two-proportion z-test
z_stat, p_value = proportions_ztest(count=successes, nobs=samples)

print(f"Z-statistic = {z_stat:.3f}")
print(f"P-value     = {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("‚Üí Reject H‚ÇÄ: Proportions are significantly different")
else:
    print("‚Üí Fail to reject H‚ÇÄ: No significant difference in proportions")
```

## 26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Two datasets (replace with your own)
a = np.array([10, 12,  9, 11, 13, 14, 12])
b = np.array([ 6,  8,  7,  9,  6,  8,  7])

# --- F‚Äëtest (two‚Äëtailed) ---
var_a, var_b = np.var(a, ddof=1), np.var(b, ddof=1)
f_stat       = var_a / var_b if var_a > var_b else var_b / var_a
dof1, dof2   = len(a) - 1, len(b) - 1
p_val        = 2 * (1 - f.cdf(f_stat, dof1, dof2))  # two‚Äësided

print(f"Variance A = {var_a:.2f}, Variance B = {var_b:.2f}")
print(f"F‚Äëstat     = {f_stat:.3f},  p‚Äëvalue = {p_val:.4f}")
if p_val < 0.05:
    print("‚Üí Reject H‚ÇÄ: variances differ.")
else:
    print("‚Üí Fail to reject H‚ÇÄ: no evidence of difference.")

# --- Visualize ---
plt.boxplot([a, b], labels=["Dataset A", "Dataset B"])
plt.title(f"F‚Äëtest for Variances\nF={f_stat:.2f}, p={p_val:.4f}")
plt.ylabel("Values")
plt.grid(True); plt.show()
```

## 27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
import numpy as np
from scipy.stats import chisquare

# Simulated observed frequencies (e.g., from a dice roll)
observed = np.array([18, 16, 20, 17, 15, 14])

# Expected frequencies assuming a fair 6-sided die
expected = np.array([100/6] * 6)  # total = 100 rolls

# Perform the Chi-square test
chi2_stat, p_val = chisquare(f_obs=observed, f_exp=expected)

# Display results
print(f"Chi-square Statistic = {chi2_stat:.2f}")
print(f"P-value               = {p_val:.4f}")

# Interpretation
if p_val < 0.05:
    print("‚Üí Reject H‚ÇÄ: Observed distribution differs significantly from expected.")
else:
    print("‚Üí Fail to reject H‚ÇÄ: No significant difference from expected.")
```
