## 2.What are the key assumptions of simple linear regression?

‚úÖ 1. Linearity
- The relationship between X and Y is linear.
- üìä The data points should follow a straight-line trend.

‚úÖ 2. Independence of Errors
- The residuals (errors) are independent of each other.
- üìà No pattern should exist in error terms across observations.

‚úÖ 3. Homoscedasticity
- The variance of errors is constant across all levels of X.
- üìâ Spread of residuals should be uniform.

‚úÖ 4. Normality of Errors
- The residuals (errors) should be normally distributed.
- Useful for valid confidence intervals and hypothesis testing.

‚úÖ 5. No Multicollinearity (only in multiple regression)
- ‚ùå Not a concern in simple regression (since only one predictor).

## 11.What is heteroscedasticity, and how does it affect the results of a multiple linear regression model?

### ‚úÖ What is Heteroscedasticity?
Heteroscedasticity means that the variance of the residuals (errors) is not constant across all levels of the independent variables.

| Problem                        | Impact                                                                                   |
| ------------------------------ | ---------------------------------------------------------------------------------------- |
| Biased standard errors         | Leads to **incorrect p-values** and confidence intervals                                 |
| Inefficient estimates          | Your model might still be unbiased, but **not optimal**                                  |
| Misleading significance tests  | Some variables may appear **significant when they‚Äôre not**, or vice versa                |
| Violates regression assumption | It breaks the **homoscedasticity assumption**, which is important for hypothesis testing |


## 12.How can you improve a multiple linear regression model with high multicollinearity?

### Multicollinearity means that two or more independent variables are highly correlated, which makes it hard for the model to estimate their individual effects accurately.

| Method                                        | Description                                                                                               |
| --------------------------------------------- | --------------------------------------------------------------------------------------------------------- |
| **1. Remove correlated predictors**           | Drop one of the highly correlated variables                                                               |
| **2. Use Principal Component Analysis (PCA)** | Transform correlated features into uncorrelated components                                                |
| **3. Combine features**                       | Create a new variable (e.g., average or ratio) from correlated ones                                       |
| **4. Regularization (Ridge/Lasso)**           | Shrinks coefficients and reduces variance; especially **Ridge Regression** is great for multicollinearity |
| **5. Check Variance Inflation Factor (VIF)**  | Remove variables with **VIF > 5 or 10** to reduce collinearity                                            |
| **6. Feature selection techniques**           | Use backward elimination or recursive feature elimination (RFE) to keep the most relevant predictors      |

## 18.What are the limitations of using R(square) as a sole measure of model      performance?

### üìä What is R¬≤?
-  R¬≤ measures the proportion of variance in the dependent variable(Y) that is explained by the independent variable(X).

     R¬≤=1-(SSR/SST)
     
### ‚ö†Ô∏è Limitations of R¬≤

1. ‚ùå High R¬≤ Doesn‚Äôt Mean Good Model
- A high R¬≤ might occur even if the model is poor (especially with non-linear    patterns, overfitting, or irrelevant variables).

2. üìà R¬≤ Always Increases With More Variables
- Adding more predictors will never decrease R¬≤, even if they‚Äôre useless.
- This can lead to overfitting ‚Äî the model fits the training data well but 
  performs poorly on new data.    
  
3. üîÆ No Insight on Causality
- R¬≤ tells you how well the model fits, not why the variables are related.
- A high R¬≤ doesn't imply X causes Y.

## 21.What does it mean if a multiple linear regression model has a high R¬≤ but low asjusted R¬≤?

- R¬≤ (Coefficient of Determination):
  Measures how much of the variance in the dependent variable is explained by the independent variables.
- Adjusted R¬≤:
  Similar to R¬≤ but penalizes for adding unnecessary predictors. It adjusts for the number of predictors and sample size.

- üìå If R¬≤ is high but Adjusted R¬≤ is low, it means:
- Your model explains a lot of variance on paper, but many predictors don‚Äôt add real explanatory power.
- Additional predictors are inflating R¬≤ artificially because R¬≤ always increases or stays the same when you add variables, even irrelevant ones.
- Adjusted R¬≤ dropped because those extra predictors do not improve the model enough to justify their inclusion.

## 23.What is polynomial regression?

### ‚úÖ Polynomial Regression ‚Äî Explained Simply
Polynomial Regression is a type of regression that models the relationship between the independent variable(s) X and the dependent variable Y as an nth-degree polynomial

## 24.How does polynomial regression differ linear regression?

### Linear Regression assumes a straight-line relationship between the independent variable(s) and the dependent variable

### Polynomial Regression models a curved relationship by including higher-order powers of the predictor:

### Complexity
- Linear Regression: Only one slope coefficient for each predictor.
- Polynomial Regression: Multiple coefficients for each power of the variable (adds flexibility but increases complexity).

###  When to Use
- Linear Regression: When the data shows a straight-line trend.
- Polynomial Regression: When the relationship is non-linear but still smooth (e.g., U-shaped or curved patterns).

## 25.When is polynomial regression used?

### ‚úÖ Common Use Cases
- 1.Curved Trends in Data
   - When a straight line (linear regression) does not fit the data well.
   - Example: Growth patterns, temperature trends, or sales data with seasonal effects.

- 2.U-shaped or Inverted U-shaped Patterns
    - Example: Advertising spend vs. sales (diminishing returns).


- 3.Modeling Complex Relationships
   - Where adding higher-order terms improves prediction accuracy without overfitting too much.

## 28.What are the limitations of polynomial regression?

### ‚úÖ 1. Risk of Overfitting
- Higher-degree polynomials can fit the training data too well, capturing noise instead of the true trend.
- Leads to poor generalization on new data.

‚úÖ 2. Sensitive to Outliers
- Outliers can heavily distort the curve because polynomial terms amplify their effect.

‚úÖ 3. Extrapolation Issues
- Predictions outside the range of training data can become wildly inaccurate because the curve can rise or fall sharply.

‚úÖ 4. High Complexity
- As the degree of the polynomial increases:
  - The model becomes harder to interpret.
  - Computational cost rises.
  - Feature space explodes (especially with multiple variables).

‚úÖ 5. Multicollinearity
- Polynomial terms (e.g., x,x2,x3) are often highly correlated, which can inflate variance in coefficient estimates.

‚úÖ 6. Assumes Global Relationship
- A single polynomial equation tries to fit the entire dataset, which may not capture local variations well.

‚úÖ 7. Not Always the Best for Non-Linear Data
- Complex non-linear patterns are sometimes better handled by:
  - Spline Regression
  - Decision Trees
  - Neural Networks
